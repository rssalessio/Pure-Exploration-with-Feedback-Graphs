```python
import numpy as np
import figure_options

from figure_options import plt, sns, rc_parameters
from baifg.utils.characteristic_time import evaluate_characteristic_time_gaussian, compute_characteristic_time_gaussian
from baifg.utils.graphs import make_symmetric_graph

fg = make_symmetric_graph()
v0 = compute_characteristic_time_gaussian(fg)[0]

xvals = np.linspace(0, 1, 50)

solutions = [evaluate_characteristic_time_gaussian(np.array([x, 0, 1-x]), fg) for x in xvals]

sns.set_style("darkgrid", rc=rc_parameters)
plt.plot(xvals, solutions)
plt.xlabel(r'$x$')
plt.ylabel(r'$T^\star$')
plt.title("Characteristic time for symmetric graph")
plt.savefig('figures/examples/characteristic_time_symmetric_graph.pdf', bbox_inches='tight')
```
